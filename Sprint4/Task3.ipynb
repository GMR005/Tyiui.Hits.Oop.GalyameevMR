```
using System;
using System.Collections.Generic;


public interface IWatchable
{
    void Watch(int views, string format);
}


public interface ICommentable
{
    void AddComment(string comment);
    void AddComment (string comment, string name);
}


public interface IMovie : IWatchable, ICommentable
{
    string GetInfo();
    string MovieQuality(int quality); 
}


public interface IMusic
{
    string GetInfo();
}


public class Movie : IMovie
{
    private string _title;
    private string _director;
    private int _year;
    private double _rating;
    private string _genre;

    public Movie(string title, string director, int year, double rating, string genre)
    {
        Title = title;
        Director = director;
        Year = year;
        Rating = rating;
        Genre = genre;
    }

    public string Title { get; set; }
    public string Director { get; set; }
    public int Year { get; set; }

    public double Rating
    {
        get { return _rating; }
        set
        {
            if (value >= 0)
            {
                _rating = Math.Round(value, 2);
            }
            else
            {
                throw new ArgumentOutOfRangeException("Рейтинг не может быть отрицательным!");
            }
        }
    }

    public string Genre { get; set; }


    public void AddComment(string comment)
    {
        Console.WriteLine($"Добавлен комментарий к фильму: {comment}");
    }
     public void AddComment(string comment, string name)
    {
        Console.WriteLine($"Добавлен комментарий к фильму от {name}: {comment}");
    }

 
    public virtual void Watch(int views, string format)
    {
        Console.WriteLine($"Просмотр: {Title}, Количество просмотров: {views}, Формат: {format}");
    }


    public virtual string GetInfo()
    {
        return $"Название: {Title}, Режиссер: {Director}, Год: {Year}, Рейтинг: {Rating}, Жанр: {Genre}";
    }

    public string MovieQuality(int quality)
    {
        string qualityMessage = $"Качество: {quality}p";
        Console.WriteLine(qualityMessage);
        return qualityMessage; 
    }
}

public class AnimatedMovie : Movie
{
    private string studio;

    public AnimatedMovie(string title, string director, int year, double rating, string genre, string studio) 
      : base(title, director, year, rating, genre)
    {
        this.studio = studio;
    }


    public override void Watch(int views, string format)
    {
        Console.WriteLine($"Просмотр мультфильма: {Title}, Количество просмотров: {views}, Формат: {format}");
    }


    public override string GetInfo()
    {
        return base.GetInfo() + $", Студия: {studio}."; 
    }
}


public class Music : IMusic
{
   private string _title;
   private string _artist;
   private int _year;

   public Music(string title, string artist, int year)
   {
       Title = title;
       Artist = artist;
       Year = year;
   }

   public string Title { get; set; }
   public string Artist { get; set; }
   public int Year { get; set; }


   public void AddComment(string comment)
   {
       Console.WriteLine($"Добавлен комментарий к музыкальному произведению: {comment}");
   }

   public virtual string GetInfo()
   {
       return $"Музыка: {Title}, Исполнитель: {Artist}, Год: {Year}.";
   }
}


public class Song : Music
{
   private double duration;

   public Song(string title, string artist, int year, double duration) 
      : base(title, artist, year)
   {
       this.duration = duration;
   }

  
   public override string GetInfo()
   {
       return base.GetInfo() + $" Длительность: {duration} минут.";
   }
}

public class GenericCollection<T>  
{
     private List<T> items = new List<T>();

     
     public void AddItem(T item)
     {
         items.Add(item);
         Console.WriteLine($"Объект добавлен в коллекцию.");
     }


     public void ShowAllItems()
     {
         foreach (var item in items)
         {
             if (item is IMovie movie) 
             {
                 Console.WriteLine(movie.GetInfo());
             }
             else if (item is IMusic music) 
             {
                 Console.WriteLine(music.GetInfo());
             }
             else 
             {
                 Console.WriteLine($"Объект типа: {item.GetType().Name}");
             }
         }
     }
}

public class MovieService
{
   private readonly IMovie _movie;


   public MovieService(IMovie movie)
   {
       _movie = movie ?? throw new ArgumentNullException(); 
   }

   public void Movie1080p()
   {
       _movie.MovieQuality(1080); 
   }

   public void Movie720p()
   {
       _movie.MovieQuality(720);  
   }
}


        var collection = new GenericCollection<IMovie>(); 

        collection.AddItem(new Movie("Властелин колец", "Питер Джексон", 2001, 9.0142,"Фэнтези"));
        collection.AddItem(new AnimatedMovie("История игрушек", "Джон Лассетер", 1995, 8.5,"Анимация", "Pixar"));
        
        var musicCollection = new GenericCollection<IMusic>(); 
        musicCollection.AddItem(new Song("Glowing Man", "Swans", 2016, 9.3));
        musicCollection.AddItem(new Song("Bodysnatchers", "Radiohead", 2007, 8.8));

        Console.WriteLine("");
        
    
        collection.ShowAllItems();


        musicCollection.ShowAllItems();

        Console.WriteLine("");

        IMovie animation = new AnimatedMovie("Аркейн", "Паскаль Шаррю", 2021, 8.5, "Анимация", "Fortiche");
        Console.WriteLine(animation.GetInfo());
        MovieService animovieService = new MovieService(animation); 
        
        animovieService.Movie1080p(); 
        
        ((ICommentable)animation).AddComment("Плохо");
        animation.AddComment("Отлично!", "Иван");

        IMovie testMovie = new Movie("Властелин колец", "Питер Джексон", 2001, 9.0142,"Фэнтези");
        
        MovieService movieService = new MovieService(testMovie); 
        
        movieService.Movie720p(); 

        Console.ReadLine();
```
